```python
import pandas as pd
import numpy as np
from patsy import dmatrices
from sklearn.model_selection import train_test_split
from sklearn import neighbors
from sklearn import metrics
```

```python
nbaRaw = pd.read_csv("/content/NBA2011_2021.csv")
```

```python
nbaRaw = nbaRaw.dropna()
```

```python
cleanNBA = nbaRaw.groupby(["Player"]).mean()
```

```python
cleanNBA
```

```python
cleanNBA = cleanNBA.drop(columns=["Rk"])
```

```python
cleanNBA.mean()
```

```python
cleanNBA["star"] = np.where(
    (cleanNBA["G"] > 41.002870) &
    (cleanNBA["GS"] > 15.990493) &
    (cleanNBA["MP"] > 848.238649 ) &
    (cleanNBA["FG"] >   4.928404) &
    (cleanNBA["FGA"] > 11.630568) &
    (cleanNBA["FG%"] >  0.424700) 

    ,
    "star"
    ,
    "lame"
)



```

```python
yearlyNBA = nbaRaw.groupby(["Player","year"]).mean()
```

```python
yearlyNBA
```

```python
yearlyNBA["yearsActive"] = yearlyNBA.groupby("Player")["Rk"].transform('count')
```

```python
yearlyNBA = yearlyNBA[yearlyNBA["yearsActive"] >= 2]
```

```python
yearlyNBA
```

```python
onlyFirst = yearlyNBA.groupby(["Player"]).head(1)
```

```python
onlyFirst
```

```python
star = cleanNBA["star"]
```

```python
combined = onlyFirst.merge(star,on = "Player",how = "left")
```

```python
combined = combined.drop(columns=["Rk","Age","yearsActive"])
```

```python
combined
```

Nearestr Neighbor

```python
combined2 = combined.copy()
```

```python
combined2['target'] = 0
```

```python
mask = combined2['star'] == "star"
combined2.loc[mask,'target'] = 1.0
```

```python
combined2 = combined2.rename(columns={"FG%":"FGPer",
                                      "3P":"P3",
                                      "3PA":"PA3",
                                      "3P%":"P3Per",
                                      "2P":"P2",
                                      "2PA":"PA2",
                                      "2P%":"P2Per"



})
```

```python
Y, X = dmatrices("target ~ 0 + G + GS + MP + FGPer + P3 + PA3 + P3Per + FGA + P2 + PA2 + P2Per + FT + FTA + ORB + DRB + TRB + AST + STL + BLK + TOV + PF + PTS",
                 combined2,return_type="dataframe")
```

```python
y = Y['target'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
model = neighbors.KNeighborsClassifier(n_neighbors=15,
                                       weights='uniform',
                                       p=2)
```

```python
model.fit(X_train, y_train)
```

```python
X_train.iloc[0:5]
```

```python
print ('Predictions =', model.predict(X_train.iloc[0:5]))
```

```python
prediction_on_training = model.predict(X_train)
```

```python
print ( metrics.accuracy_score(y_train, prediction_on_training))
```

```python
model.predict_proba(X_train.iloc[0:5])
```

```python
model.predict(X_train.iloc[0:5])
```

```python
predicted_classes = model.predict(X_test)
print (metrics.accuracy_score(y_test, predicted_classes))
```
